## **`Vector Operations`**


```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

### `Plotting a vector`

```python

plt.quiver(10,21,11,41)


```

```python
# Create a quiver plot
plt.quiver(0,0,4,5,scale_units = "xy",angles="xy",scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)


# Display the plot
plt.show()
```

```python
# Create a quiver plot
plt.quiver(0,0,4,5,scale_units = "xy",angles="xy",scale=1,color="blue")
plt.xlim(-8,8)
plt.ylim(-8,8)


# Display the plot
plt.show()
```

```python
#multiple vectors 
# Create a quiver plot
plt.quiver(0,0,4,5,scale_units = "xy",angles="xy",scale=1,color="b")
plt.quiver(0,0,-3,-2,scale_units = "xy",angles="xy",scale=1,color="orange")

plt.xlim(-8,8)
plt.ylim(-8,8)


# Display the plot
plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])

sum_vector = vector_1 + vector_2


print(sum_vector)
```

```python
plt.quiver(0,0,2,3,scale_units = "xy",angles="xy",scale=1,color="b")
plt.quiver(0,0,3,-2,scale_units = "xy",angles="xy",scale=1,color="orange")
plt.quiver(3,-2,2,3,scale_units = "xy",angles="xy",scale=1,color="g")
plt.quiver(2,3,3,-2,scale_units = "xy",angles="xy",scale=1,color="orange")


plt.quiver(0,0,5,1,scale_units = "xy",angles="xy",scale=1,color="red")


plt.xlim(-8,8)
plt.ylim(-8,8)


# Display the plot
plt.show()
```
