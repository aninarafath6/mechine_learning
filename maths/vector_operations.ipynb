## **`Dot Product of vectors`**
**both vectors must have the same** The resultant of  dot product of vectors should be the scalar 

${\vec a.\vec b} =  {|\vec a ||\vec b|Cos\Theta}$

${cos \Theta} = { \frac{\vec a.\vec b}{|\vec a ||\vec b|} }$


$Magnitude of  vector (| v|) = {\sqrt{(v_x^2 + v_y^2)}}$



```python
vector_1 = np.array([2,3,8])
vector_2 =  np.array([4,4,9])


dot_product = vector_1.dot(vector_2)
dot_product
```

## **`Cross Product of vectors`**
${\vec a\times\vec b} =  {|\vec a ||\vec b|Sin\Theta}$

${Sin \Theta} = { \frac{\vec a\times\vec b}{|\vec a ||\vec b|} }$


$Magnitude of  vector (| v|) = {\sqrt{(v_x^2 + v_y^2)}}$



```python
cross_product = np.cross(vector_1,vector_2)
cross_product
```
