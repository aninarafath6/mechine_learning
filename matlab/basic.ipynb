# **`Matplotlib Pyplot`**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_points = np.array([1,8])
y_points = np.array([3,10])
plt.plot(x_points, y_points,marker="o")
plt.show()
```

```python
y_points = np.array([10,20,2,10,40,20,22,30])
plt.plot(y_points,marker="o")
plt.show()
```

# Shorter Syntax

 `The line style can be written in a shorter syntax:`

**linestyle can be written as ls.**</br>

dotted can be written as :.</br>
dashed can be written as --.

```python
# styling 
y_points = np.random.randint(5,100,40)

plt.plot(y_points,marker="o",linestyle="--") # == 
plt.show()
```

```python
plt.plot(y_points,marker="o",ls=":")
plt.show()
```

```python
plt.plot(y_points,marker="o",ls="-.",color="green") # dash dotted
plt.show()
```

## **`Line Width`**

### You can use the keyword argument linewidth or the shorter lw to change the width of the line.
#### The value is a floating number, in points:

```python
plt.plot(y_points,marker="o",lw=4,color="red")
plt.show()
```

## **`Multiple Lines`**
### You can plot as many lines as you like by simply adding more plt.plot() functions:


```python
y1_points = np.random.randint(1,200,10)
y2_points = np.random.randint(1,200,10)

plt.plot(y1_points,marker='o')
plt.plot(y2_points,marker="0")


plt.show()
```
