## **`Seaborn`**
`Data Visualization Library`

```python
# Essential Packages
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Seaborn has some built in data sets

#total bill vs tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",col="time",hue="smoker",size="size",style="sex")
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.value_counts("species")
```

```python
sns.scatterplot(x="sepal_length",y="sepal_width",hue="species",data=iris)
```

```python
sns.scatterplot(x="sepal_length",y="petal_width",hue="species",data=iris)
```

```python
# Load titanic data set
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# Count plot
sns.countplot(x="class",data=titanic)
titanic.value_counts("class")
```

```python
sns.countplot(x="survived",data=titanic)
titanic.value_counts("survived")
```

```python
# Bar chart
sns.barplot(x="sex",y="survived",hue="class",data=titanic)
```
