```python
#essential packages 
import requests
from bs4 import BeautifulSoup
import json
```

```python
# URL of the file containing the results of the analysis
# access the  web content 
URL = "https://www.isro.gov.in/SpacecraftMissions.html#"
response = requests.get(URL)
```

```python
soup = BeautifulSoup(response.text,'lxml')
spacecrafts = soup.find_all("a", {"class":"out"})
```

```python

data = []
for spacecraft in spacecrafts:
        title = spacecraft.text
        link = spacecraft["href"]
        url = "https://www.isro.gov.in/{craft}".format(craft=link)
        response = requests.get(url)
        soup = BeautifulSoup(response.text,'lxml')
        tables = soup.find_all("table", {"class":"pContent table table-striped table-bordered"})
        count = 0
        for table in tables:
                rows = table.find_all('tr')
                dt = {"name":title,"id":1}
                for row in rows:
                        cols = row.find_all('td')
                        cols = [ele.text.strip() for ele in cols]
                        if(len(cols)== 2):
                                dt["".join("_".join(cols[0].lower().split(" ")).split("\n"))] = (" ".join((cols[1]).split("\n"))).strip()

                data.append(dt)



# print(data)
for i in range(len(data)):
        data[i]["id"] = (len(data)-i)
json_data =  json.dumps(data,indent=2)
print(json_data)




```
