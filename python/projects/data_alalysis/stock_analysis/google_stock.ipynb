#### **`Data Visualization`**
Next, we will use various techniques to visualize the data, including line charts, candlestick charts, and scatterplots. These visualizations will help us identify patterns in the data and make informed investment decisions.



```python
# sns.lineplot(data = google,x="date",y="open") # seaborn method
plt.subplot(1 ,2,1)
google["open"].plot(label=" Open Price",figsize=(15,7),)
google["close"].plot(label=" Open Price")
google["high"].plot(label=" High Price")
google["low"].plot(label=" Low Price")

font1 = {'family':'sans','color':'red','size':15}
plt.suptitle("Google Stock",fontdict=font1)


plt.grid()
plt.ylabel("Stock Price")
plt.xticks(rotation=45)
plt.title("Stock PricemAnalysis")


plt.subplot(1 ,2,2)
google["volume"].plot(label=" Volume Price",figsize=(15,7),)
plt.title("Volume Analysis")
plt.ylabel("Volume")

plt.xticks(rotation=45)

plt.grid()
plt.show()
```
